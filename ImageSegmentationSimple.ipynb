```python
"""
Implementation of Thresholding based segmentation
In global thresholding, we used an arbitrary chosen value as a threshold. 
In contrast, Otsu's method avoids having to choose a value
and determines it automatically."""

import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
image = cv2.imread("fl.jpg", 0)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
flattened_image = image.reshape((image.shape[0] * image.shape[1],))
```

```python
flattened_image.shape
```

```python
plt.figure()
sns.distplot(flattened_image, kde=True)
plt.show()
```

```python
"""Applying Otsu Thresholding to the image for segmentation"""
ret, thresh1 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print("Otsu threshold: ", thresh1)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(thresh1, cmap="binary")
plt.axis("off")
plt.show()
```
